```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\adead\Downloads\products_wk9_thurs.csv")
df1.head()
```

```python
df2 = pd.read_csv(r"C:\Users\adead\Downloads\customers_wk9_thurs.csv")
df2.head()
```

```python
df3 = pd.read_csv(r"C:\Users\adead\Downloads\sales_wk9_thurs.csv")
df3.head()
```

```python
df1['launch_date'] = pd.to_datetime(df1['launch_date'])
```

```python
df2['registration_date'] = pd.to_datetime(df2['registration_date'])
```

```python
df3['transaction_date'] = pd.to_datetime(df3['transaction_date'])
```

```python
df3['year'] = df3['transaction_date'].dt.year
df3['month'] = df3['transaction_date'].dt.month
df3['day_of_week'] = df3['transaction_date'].dt.day_name()
```

```python
earliest_date = df3['transaction_date'].min()
latest_date = df3['transaction_date'].max()
```

```python
df3[['transaction_date', 'year', 'month', 'day_of_week']].head()
```

```python
print(f"Sales data starts on: {earliest_date}")
print(f"Sales data ends on: {latest_date}")
```

```python
busiest_day = df3['day_of_week'].value_counts()
print("Transactions by Day of Week:")
print(busiest_day)
```

```python
busiest_month = df3['month'].value_counts()
print("\nTransactions by Month:")
print(busiest_month)
```

```python
sales_per_year = df3['year'].value_counts().sort_index()
print("\nTransactions per Year:")
print(sales_per_year)
```

```python
avg_price_per_category = df1.groupby('category')['price'].mean()
print("Average Price per Category:")
print(avg_price_per_category)
```

```python
products_per_category = df1['category'].value_counts()
print("\nProduct Count per Category:")
print(products_per_category)
```

```python
avg_age_state = df2.groupby('state')['age'].mean()
print("Average Age per State:")
print(avg_age_state)
```

```python
customer_sales = pd.merge(df3, df2, on='customer_id')
```

```python
total_spending = customer_sales.groupby(['first_name', 'last_name'])['total_amount'].sum().sort_values(ascending=False)

print("Total Spending per Customer (Top 5):")
print(total_spending.head())
```

```python
print(customer_sales.columns)
```

```python
purchase_counts = customer_sales.groupby(['first_name', 'last_name']).size().sort_values(ascending=False)

print("\nMost Frequent Shoppers (Top 5):")
print(purchase_counts.head())
```

```python
revenue_by_month = customer_sales.groupby('month')['total_amount'].sum()
print("Total Revenue by Month:")
print(revenue_by_month)
```

```python
avg_trans_day = customer_sales.groupby('day_of_week')['total_amount'].mean()
print("\nAverage Transaction Amount by Day of Week:")
print(avg_trans_day)
```

```python
sales_with_products = pd.merge(customer_sales, df1[['product_id', 'category']], on='product_id')
```

```python
category_revenue = sales_with_products.groupby('category')['total_amount'].sum().sort_values(ascending=False)

print("\nTotal Revenue per Category:")
print(category_revenue)
```

```python
top_category = category_revenue.idxmax()
print(f"\nThe product category generating the most revenue is: {top_category}")
```

```python
customers_per_state = df2['state'].value_counts()
print("Customers per State:")
print(customers_per_state)
```

```python
top_state = customers_per_state.idxmax()
print(f"\nThe state with the most customers is: {top_state}")
```

```python
top_5_spenders = customer_sales.groupby(['first_name', 'last_name'])['total_amount'].sum().sort_values(ascending=False).head(5)

print("Top 5 Customers by Total Spending:")
print(top_5_spenders)
```

```python
best_day = avg_trans_day.idxmax()
max_val = avg_trans_day.max()

print(f"The day with the highest average transaction value is {best_day} with an average of ${max_val:.2f}")
```
