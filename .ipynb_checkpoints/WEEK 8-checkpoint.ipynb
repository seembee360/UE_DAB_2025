```python
df_dict = {
    "Inventory": df1, 
    "Sales": df2, 
    "Stores": df3, 
    "Products": df4, 
    "Customers": df5
}

for name, df in df_dict.items():
    print(f"--- Missing Values in {name} ---")
    print(df.isnull().sum())
    print("\n")
 
```

```python
for name, df in df_dict.items():
    print(f"--- Data Types in {name} ---")
    print(df.dtypes)
    print("\n")
```

```python
unique_products = df4['product_id'].nunique()
print(f"Total unique products: {unique_products}")
```

```python
top_5_expensive = df4.sort_values(by='price', ascending=False).head(5)
print("Top 5 most expensive products:")
display(top_5_expensive[['product_name', 'price']])
```

```python
largest_store_idx = df3['size_sqft'].idxmax()
largest_store = df3.loc[largest_store_idx]
print(f"The largest store is {largest_store['store_name']} with {largest_store['size_sqft']} sqft.")
```

```python
customer_distribution = df5['state'].value_counts()
print("Customer distribution by state:")
print(customer_distribution)
```

```python
df3['size_sqft'] = df3['size_sqft'].fillna(df3['size_sqft'].mean())
df1['reorder_level'] = df1['reorder_level'].fillna(df1['reorder_level'].mean())
```

```python
df5['email'] = df5['email'].fillna(df5['email'].mode()[0])
```

```python
df1['last_restock_date'] = df1['last_restock_date'].ffill()
df5['registration_date'] = df5['registration_date'].ffill()
```

```python
print("Missing values in Stores:", df3['size_sqft'].isnull().sum())
print("Missing values in Inventory:", df1['reorder_level'].isnull().sum())
```

```python
print(f"Duplicate rows in Customers: {df5.duplicated().sum()}")
print(f"Duplicate rows in Products: {df4.duplicated().sum()}")
```

```python
df5 = df5.drop_duplicates(keep='first')
df4 = df4.drop_duplicates(keep='first')
```

```python
print(f"\nNew Customers count: {len(df5)}")
print(f"New Products count: {len(df4)}")
```
